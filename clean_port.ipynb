```python
import os
import signal
import subprocess
import platform
```

```python
def find_python_server_processes(port):
    try:
        # List all processes and find Python processes using the specified port
        if platform.system() == "Windows":
            # Use netstat to find processes using the given port
            result = subprocess.run(['netstat', '-ano'], capture_output=True, text=True)
            lines = result.stdout.splitlines()
            
            for line in lines:
                if f":{port}" in line and 'LISTEN' in line:
                    # Extract the PID
                    parts = line.split()
                    pid = parts[-1]
                    # Check if the PID corresponds to a Python process
                    pid_process = subprocess.run(['tasklist', '/FI', f'PID eq {pid}', '/FI', 'IMAGENAME eq python.exe'], capture_output=True, text=True)
                    if 'python.exe' in pid_process.stdout:
                        print(f"Killing Python server process with PID: {pid}")
                        # Use taskkill command to forcefully kill the process
                        subprocess.run(['taskkill', '/PID', pid, '/F'])
        
        elif platform.system() in ["Linux", "Darwin"]:  # Linux or macOS
            # Use lsof to find the process using the given port
            result = subprocess.run(['lsof', '-i', f':{port}'], capture_output=True, text=True)
            lines = result.stdout.splitlines()
            for line in lines:
                if 'python' in line:
                    parts = line.split()
                    pid = parts[1]  # PID is the second column in lsof output
                    print(f"Killing Python server process with PID: {pid}")
                    os.kill(int(pid), signal.SIGKILL)

    except Exception as e:
        print(f"Error occurred: {e}")
```

```python
ports_to_clean = [8081, 8082, 8083]
for port in ports_to_clean:
    find_python_server_processes(port)
```
